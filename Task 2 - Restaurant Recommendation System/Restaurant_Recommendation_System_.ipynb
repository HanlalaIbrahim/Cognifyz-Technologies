```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
dataset = pd.read_csv('Dataset .csv')
dataset.head()
```

```python
dataset['Cuisines'] = dataset['Cuisines'].fillna('')
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(dataset['Cuisines'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_cuisine_recommendations(cuisine, cosine_sim=cosine_sim):
    indices = dataset[dataset['Cuisines'].str.contains(cuisine, case=False)].index

    if len(indices) == 0:
        return f"No restaurants found for the cuisine: {cuisine}"

    sim_scores = sum(cosine_sim[indices, :])
    sim_scores = sorted(list(enumerate(sim_scores)), key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[:10]
    restaurant_indices = [i[0] for i in sim_scores]

    return dataset[['Restaurant Name', 'Cuisines', 'City', 'Aggregate rating']].iloc[restaurant_indices].reset_index(drop=True)
```

```python
cuisine_recommendations = get_cuisine_recommendations('Italian')
print("\nRecommended Restaurants based on cuisine 'Italian':\n")
cuisine_recommendations
```
